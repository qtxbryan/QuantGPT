```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
initial_value = {}
initial_investment = 10000
weights = [0.6, 0.4]
```

```python
for i, x in enumerate(['AAPL', 'AMZN']):
    initial_value[x] = [initial_investment * weights[i]]
    
initial_value
```

```python
data = yf.download(['AAPL', 'AMZN'], start='2024-12-01', end='2024-12-31')
```

```python
returns = data.pct_change().dropna()
returns
```

```python
dates_list = [str(x.date()) for x in returns.cumsum().index]
dates_list
```

```python
constitutents_breakdown = pd.DataFrame(
    (np.array(returns.cumsum()) + 1) * np.array(pd.DataFrame(initial_value)),
    index=dates_list,
    columns=returns.cumsum().columns
)
constitutents_breakdown
```
